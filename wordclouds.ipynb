# Word Cloud generation

In this notebook we generate Word Clouds based on the most frequent words in the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud
```

```python
pd.set_option('display.max_colwidth',None)
pd.set_option('display.float_format',lambda x: '%.2f' % x)
```

Let's import the preprocessed dataset.

```python
df_full = pd.read_pickle('preprocessed_dataset.pkl')
```

```python
df_full.info()
```

In the following we create dictionaries which keep count of the frequency of words in several text columns of the dataset. Based on these, we create the Word Clouds.

```python
names = []

for name in df_full.plain_name:
    names += name

names_dict = Counter(names)
```

```python
fig = plt.figure(figsize=(12,10))
wordcloud = WordCloud().generate_from_frequencies(names_dict)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.savefig('names_wordcloud.png',bbox_inches='tight')
plt.show()
```

```python
descriptions = []

for description in df_full.plain_description:
    descriptions += description

descriptions_dict = Counter(descriptions)
```

```python
fig = plt.figure(figsize=(12,10))
wordcloud = WordCloud().generate_from_frequencies(descriptions_dict)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
fig = plt.figure(figsize=(12,10))
wordcloud = WordCloud().generate_from_frequencies(ingredients_dict)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
tags = []

for tag in df_full.plain_tags:
    tags += tag

tags_dict = Counter(tags)
```

```python
fig = plt.figure(figsize=(12,10))
wordcloud = WordCloud().generate_from_frequencies(tags_dict)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```
